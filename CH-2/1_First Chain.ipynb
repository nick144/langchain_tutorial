```python
from langchain_openai import ChatOpenAI
import os
from dotenv import load_dotenv
from pydantic import BaseModel
from langchain_core.output_parsers import StrOutputParser

from langchain_core.prompts import ChatPromptTemplate, PromptTemplate

load_dotenv()

if "OPENAI_API_KEY" not in os.environ:
    raise ValueError("OPENAI_API_KEY environment variable is not set.")

```

```python
# Task 1 [Prompt]
# Prompt
Prompt_template = ChatPromptTemplate.from_messages(
    [
        ("system", "You are a helpful assistant."),
        ("human", "{input}"),
    ]
)
```

```python
# Task 2 [LLM]
llm_openai = ChatOpenAI(model="gpt-5-nano", temperature=0)
```

```python
# Task 3 [Parser]
str_parser = StrOutputParser()
```

# **Manual Invocation**

```python
template = PromptTemplate.format_prompt(input="What is the capital of France?").to_messages()

result = llm_openai.invoke(template)

final_result = result.content
```

# **Chain Invocation**

```python
chain = Prompt_template | llm_openai | str_parser

chain_result = chain.invoke({"input": "What is the capital of France?"})
```

```python
from langchain_core.runnables import RunnableSequence

sequence = RunnableSequence([Prompt_template, llm_openai, str_parser])
sequence_result = sequence.invoke({"input": "What is the capital of France?"})
```
