```python
from langchain_core.messages import HumanMessage, SystemMessage, AIMessage
from langchain_openai import ChatOpenAI

llm_model = ChatOpenAI(model="gpt-5-mini", temperature=0)

my_messages = [
    SystemMessage(content="You are a gen-z assistant who always answers in a fun way."),
    HumanMessage(content="Bro, tell me a fun fact about space!"),
]

llm_model.invoke(my_messages).content
```

```python
from langchain_core.prompts import ChatPromptTemplate

chat_prompt = ChatPromptTemplate.from_messages(
    [
        # ("system", "You are a gen-z assistant who always answers in a fun way."),
        ("system", "You are a {tone} assistant."),
        ("user", "Tell me a fun fact about {topic}!"),
        # SystemMessage(content="You are a gen-z assistant who always answers in a fun way."),
        # HumanMessage(content="Tell me a fun fact about {topic}!"),
    ]
)

# User Input
UserInput = input("Topic of fun fact: ")
userTone = input("Tone of the response: ")

""" Prompt Ready to be sent to the LLM """
ready_prompt = chat_prompt.invoke({"topic": UserInput, "tone": userTone})

""" Prompt checked Response """
print(ready_prompt.messages)
# llm_model.invoke(ready_prompt).content

""" Final Response from the LLM """
llm_model.invoke(ready_prompt.messages).content
```
