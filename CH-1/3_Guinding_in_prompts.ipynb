# ***Guiding in prompts***

```python
from langchain_openai import ChatOpenAI
import os
from dotenv import load_dotenv

load_dotenv()

if os.getenv("OPENAI_API_KEY") is None:
    raise ValueError("OPENAI_API_KEY environment variable not set")
```

```python
from langchain_core.prompts import PromptTemplate

llm_model = ChatOpenAI(model="gpt-5-nano", temperature=0)
# llm_model.invoke(ready_prompt).content
```

```python
from pydantic import BaseModel, Field

class Joke(BaseModel):
    setup: str = Field(description="The setup of the joke")
    punchline: str = Field(description="The punchline of the joke")

```

```python
llm_structured_response = llm_model.with_structured_output(Joke)

ready_prompt = PromptTemplate.from_template("Tell me a joke. Generate Output in JSON format with key-value pair format with setup and punchline.").invoke({})

response = llm_structured_response.invoke(ready_prompt)
```
